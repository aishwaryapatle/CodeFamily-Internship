

# **Data Visualization**



> **Univariate Analysis**

> Histogram

```python
# Histogram
data.hist(figsize=(15,10), layout=(2,4), color="indigo");
```



> Box Plot



```python
# Box Plot
sns.set(style="whitegrid")
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x='Species',y='SepalLengthCm',data=data)
plt.subplot(2,2,2)
sns.boxplot(x='Species',y='SepalWidthCm',data=data)
plt.subplot(2,2,3)
sns.boxplot(x='Species',y='PetalLengthCm',data=data)
plt.subplot(2,2,4)
sns.boxplot(x='Species',y='PetalWidthCm',data=data)
plt.show()
```



> Violin Plot



```python
# Violin Plot
sns.set(style="whitegrid")
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.violinplot(x='Species',y='SepalLengthCm',data=data)
plt.subplot(2,2,2)
sns.violinplot(x='Species',y='SepalWidthCm',data=data)
plt.subplot(2,2,3)
sns.violinplot(x='Species',y='PetalLengthCm',data=data)
plt.subplot(2,2,4)
sns.violinplot(x='Species',y='PetalWidthCm',data=data)
plt.show()
```

> Scatter Plot

```python
# Scatter Plot
sns.FacetGrid(data, hue="Species", height=5).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
sns.FacetGrid(data, hue="Species", height=5).map(plt.scatter,'SepalLengthCm','PetalWidthCm').add_legend()
sns.FacetGrid(data, hue="Species", height=5).map(plt.scatter,'PetalLengthCm','PetalWidthCm').add_legend()
sns.FacetGrid(data, hue="Species", height=5).map(plt.scatter,'PetalLengthCm','SepalWidthCm').add_legend()
```

> **Multivariate Analysis**

> Pair Plot

```python
# Pair plot gives the realtionship b/w all features distribution with each other..
sns.pairplot(data=data, hue='Species')
plt.show()
```

> Correlational Matrix

```python
# Correational matrix
sns.heatmap(data.corr(), annot=True);
```
